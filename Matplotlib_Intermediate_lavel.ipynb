```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
data = pd.DataFrame(np.random.randn(20, 8), columns=['A', 'B', 'C', 'D', 'E','F','G','H'])

correlation_matrix = data.corr()

plt.figure(figsize=(10, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid data
x = np.linspace(1, 50,25)
y = np.linspace(1, 50,25)
x, y = np.meshgrid(x, y)

# Compute Z values (for example, Z = x^2 + y^2)
z = np.sqrt(x**2 + y**2)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
ax.plot_surface(x, y, z, cmap='plasma',edgecolor='black')

# Show plot
plt.show()
```

```python
x = np.linspace(1, 5, 3)
y = np.linspace(1, 5, 3)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis',edgecolor='black')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z = f(x, y)")
ax.set_title("3D Surface Plot")

plt.show()
```

```python
# Animation graph
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

def update(frame):
    line.set_ydata(np.sin(x + frame / 10))
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50, blit=True)

plt.show()
```

```python
#Multiple Axes Create
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Plot on the first axis
ax1.plot(x, y1, 'r-', label='sin(x)')
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.legend()

# Plot on the second axis
ax2.plot(x, y2, 'b-', label='cos(x)')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.legend()

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
#Sharing Axes:
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create two subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True)  # 2 rows, 1 column

# Plot on the first axis
ax1.plot(x, y1, 'r-', label='sin(x)')
ax1.set_title('Sine Wave')
ax1.set_ylabel('sin(x)')
ax1.legend()

# Plot on the second axis
ax2.plot(x, y2, 'b-', label='cos(x)')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.legend()

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
